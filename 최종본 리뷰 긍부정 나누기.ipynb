```python
import pandas as pd
from pandas import DataFrame
from pandas import Series
import numpy as np
import re
from konlpy.tag import Mecab
import pandas as pd
import matplotlib as mpl
from gensim.models import Word2Vec
import numpy as np
from konlpy.tag import Okt
from tqdm import tqdm
import os
okt = Okt()
```

```python
#긍정 데이터프레임 만들기(상품이름,리뷰 키워드 추출 후 저장)
os.chdir("C:/Users/won/teamproject")
prod_name_list=[]
for a in range(0,10):
    fileName="pos_sunitem"+str(a)+"keywords"
    
    whole_word_list=[]
    
    pos_df = pd.read_csv("sunitem"+str(a)+"pos_sunscreen.csv",encoding="cp949")
    prod_name_list.append(pos_df.loc[1]['상품이름'])
    
    
    text=''
    review=[]
    for each_review in pos_df['리뷰']:
        each_review = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "", each_review)
        review.append(each_review)

    pos_df = pos_df.drop(['리뷰'], axis=1)
    pos_df['review'] = review

    #명사 추출 과정
    pos_keywords_list=[]
    for i in range(0,len(pos_df['review'])):

        pos_keyword = okt.nouns(pos_df['review'][i])

        pos_keywords_list.extend(pos_keyword)
        whole_word_list.extend(pos_keyword)

    df_word_count = pd.DataFrame()
    
    word_cloud_dict = {}

    word_list = list(set(pos_keywords_list))
    
    word_cnt_list = []
    word_list_1=[]

    for word in tqdm(word_list):

        if len(word) != 1:
            word_list_1.append(word)
            word_cnt = pos_keywords_list.count(word)
            word_cnt_list.append(word_cnt)
            word_cloud_dict[word] = word_cnt

        else:
            pass
    
    df_word_count['word'] = word_list_1
    df_word_count['word_cnt'] =  word_cnt_list
    df_word_count = df_word_count.sort_values('word_cnt',ascending=False)
    df_word_count.to_csv('C:/Users/won/teamproject/keywords/postive/'+fileName+'.csv',index=False, encoding='cp949')
```
